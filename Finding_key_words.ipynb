```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
  
  
all_descriptions = '' 
  
for i in response_json['articles']: 
      
    if i['description'] != None: 
        all_descriptions += i['description'] + ' '
        
for i in response_json['articles']: 
      
    if i['title'] != None: 
        print(i['title'])
          
wordcount={} 
for word in all_descriptions.split(): 
    if word not in wordcount: 
        wordcount[word] = 1
    else: 
        wordcount[word] += 1
  
for k,v, in sorted(wordcount.items(), 
                   key=lambda words: words[1],  
                   reverse = True): 
    print(k,v)
```

```python
import nltk
import nltk.corpus
from nltk.corpus import stopwords
unessecary_words = stopwords.words('english')
unessecary_words
```

```python
r = all_descriptions.replace('\s+',' ').replace(',', ' ').replace('.', ' ').replace('<ol>',' ').replace('<li>',' ').replace('</li>',' ')
words = r.split() 
rst = [word for word in words if 
       ( word.lower() not in unessecary_words  
        and len(word) > 3) ] 
  
rst = ' '.join(rst) 
   
wordcount={} 
  
for word in rst.split(): 
      
    if word not in wordcount: 
        wordcount[word] = 1
    else: 
        wordcount[word] += 1
   
#for k,v, in sorted(wordcount.items(), 
                   #key=lambda words: words[1], 
                   #reverse = True): 
    #print(k,v)
for_the_cloud = {}
cnt = 10
for k,v, in sorted(wordcount.items(), 
                   key=lambda words: words[1], 
                   reverse = True): 
    if cnt == 0:
        break
    for_the_cloud[k]=v
    cnt = cnt - 1
print(for_the_cloud.items())
for k,v, in sorted(wordcount.items(), 
                   key=lambda words: words[1], 
                   reverse = True):
    print(k,v)
```
